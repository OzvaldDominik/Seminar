```python
import sklearn
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np
import matplotlib.pyplot as plt
import pgmpy as pgm
%pylab inline
from scipy.stats import norm
```

```python
x_range=np.linspace(-12,12,150)
x1_pdf=norm.pdf(x_range,-5,np.sqrt(5))
x2_pdf=norm.pdf(x_range,0,1)
x3_pdf=norm.pdf(x_range,5,np.sqrt(10))

plt.plot(x_range,x1_pdf,label="y = 1")
plt.plot(x_range,x2_pdf,label="y = 2")
plt.plot(x_range,x3_pdf,label="y = 3")

plt.title("p(x|y)")
plt.xlabel("x")
plt.legend()
plt.grid()

plt.show()
```

```python
p_xy1=x1_pdf*0.3
p_xy2=x2_pdf*0.2
p_xy3=x3_pdf*0.5

plt.plot(x_range,p_xy1,label="y = 1")
plt.plot(x_range,p_xy2,label="y = 2")
plt.plot(x_range,p_xy3,label="y = 3")

plt.xlabel("x")
plt.title("p(x,y)")
plt.legend()
plt.grid()

plt.show()
```

```python
p_x=p_xy1+p_xy2+p_xy3

plt.plot(x_range,p_x)

plt.xlabel("x")
plt.title("p(x)")
plt.grid()

plt.show()
```

```python

plt.plot(x_range[50:-50],(p_xy1/p_x)[50:-50],label="y = 1")
plt.plot(x_range[50:-50],(p_xy2/p_x) [50:-50],label="y = 2")
plt.plot(x_range[50:-50],(p_xy3/p_x) [50:-50],label="y = 3")

plt.xlabel("x")
plt.title("p(x,y)")
plt.legend()
plt.grid()

plt.show()
```
